```python
import numpy as np
import plotly.graph_objects as go 
import scipy.stats as stats
```

```python
n, p = 100000, .5

W_list = []

for i in range(0, 100000):
    H = np.random.binomial(n, p)
    W = (2 * H - n) ** 2 / n
    W_list.append(W)
```

```python
plot_chi = go.Figure()
hist_chi = go.Histogram(x = W_list, name="Sample", histnorm = 'probability density', nbinsx = 250)
plot_chi.update_layout(xaxis_range=[0,10], yaxis_range=[0,2])
plot_chi.add_trace(hist_chi)

x_ = np.linspace(0,30,10000)
fx = [stats.chi2.pdf(xi, 1) for xi in x_]

graph_chi = go.Scatter(x=x_, y=fx, mode='lines', name="Theoretical PDF of Chi_2(1)")
plot_chi.add_trace(graph_chi)
plot_chi.show()
```
